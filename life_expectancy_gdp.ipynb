```python
df = pd.read_csv('all_data.csv')
df = df.rename({"Life expectancy at birth (years)":"LifeExpt"}, axis='columns')
print('Data Shape:' + str(df.shape))
df.head()
```

```python
sns.histplot(df.GDP)
```
